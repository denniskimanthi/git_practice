# **VISUALIZATION**

**IDENTIFYING THE MOST POPULAR MOVIE GENRE IN THE DATA**

```python
#define a variable that returns value counts of the genres
genre_counts = merged_data['genres'].value_counts()
genre_counts.head(10)
```

**Visualizing the most popular movie genre**

```python
#display bar graph of popular movie genre
#display top 10 popular movie genres
top_genres = genre_counts.head(10)
# Plotting
plt.figure(figsize=(10, 6))  # Adjust the figure size
plt.bar(x=top_genres.index, height=top_genres.values, color='skyblue')

# Adding titles and labels
plt.title('Popular Movie Genres')
plt.xlabel('Genres')
plt.ylabel('Distribution')

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better appearance
plt.show()
```

From the bar graph above we see that the most popular movie genre is Drama,followed by Documentary then Horror and so forth

**VISUALIZING THE VOTE_AVERAGE DISTRIBUTION**

The code below is meant to visualize the Vote_average of all genres across the dataset

```python
#Display a variable that returns value counts of Vote_average
vote_average = merged_data['vote_average rating'].value_counts()
vote_average.head(10)
```

Now we plot the above values into a bar graph

```python
#create a variable 'vote_rating in our data
vote_rating = merged_data['vote_average rating']
#Create histogram graph
plt.figure(figsize=(10, 6))
plt.hist(vote_rating, bins=10, edgecolor='black')

# Add titles and labels
plt.title('Vote average Distribution')
plt.xlabel('Vote average rating')
plt.ylabel('Distribution')

# Display the plot
plt.show()

```

From the visualization above we notice that the graph is skewed to the right meaning that most viewers voted for high rated movies and few voted for low rated movies. This means that most viewers preferd to watch high rated films
